```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("readCsvFile").getOrCreate()
Salesdf = spark.read.csv("Sales.csv",header=True,inferSchema=True)
Salesdf.show()
Salesdf.count()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("readCsvFile").getOrCreate()
Productsdf = spark.read.csv("ProductsNew.csv",header=True,inferSchema=True)
Productsdf.show()
Productsdf.count()
```

```python
joindf = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey)
joindf.show()
joindf.count()
```

```python
joindf = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"inner")
joindf.show()
joindf.count()
```

```python
joindf = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"Left")
joindf.show()
joindf.count()
```

```python
joindf = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"leftanti")
joindf.show(joindf.count(),truncate=False)
```

```python
joindf = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"leftSemi")
joindf.show(joindf.count(),truncate=False)
```

```python
customerdf = spark.read.csv("Customers.csv",header=True,inferSchema=True)
customerdf.show()
```

```python
joindf2 = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"inner").join(customerdf,Salesdf.CustomerKey==customerdf.CustomerKey)
joindf2.show()
```

```python
joindf4 = Salesdf.join(Productsdf,Salesdf.ProductKey==Productsdf.ProductKey,"inner").drop(Salesdf.ProductKey)
joindf4.show()
```

```python
SalesMultidf = spark.read.csv("SalesMulti.csv",header=True,inferSchema=True)
SalesMultidf.show()
```

```python
ProductsMultidf = spark.read.csv("ProductsMulti.csv",header=True,inferSchema=True)
ProductsMultidf.show()
```

```python
joinmultidf = SalesMultidf.join(ProductsMultidf,["ProductId","Country"])
joinmultidf.show()
```

```python
productmultidf2 = ProductsMultidf.withColumnRenamed("ProductId","ProductKey")
productmultidf2.show()
```

```python
joinmultidf2 = SalesMultidf.join(productmultidf2,(SalesMultidf.ProductId == productmultidf2.ProductKey) & (SalesMultidf.Country == productmultidf2.Country),"inner").drop("ProductKey",productmultidf2.Country)
joinmultidf2.show()                                 
```
