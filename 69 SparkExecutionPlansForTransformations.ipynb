```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("readCsvFile").getOrCreate()
df = spark.read.csv("Products.csv",header=True,inferSchema=True)
df.show()
df.count()
```

```python
df.explain()
```

```python
df.explain(True)
```

```python
df2 = df.filter("Country = 'India' ")
df2.show()
df2.count()
```

```python
df2.explain(True)
```

```python
from pyspark.sql.functions import *
```

```python
df3 = df.withColumn("CountryOfOrgin",lit("India"))
df3.show()
df3.count()
```

```python
df3.explain(True)
```

```python
df4 = df.withColumn("OrginOfCountry",lit("India")).filter("Country = 'France'")
df4.show()
df4.count()
```

```python
df4.explain(True)
```

```python
df5 = df.select(sum("SalesAmount").alias("TotalSales"))
df5.show()
```

```python
df5.explain(True)
```

```python
df6 = df.groupBy("Country").sum("SalesAmount")
df6.show()
```

```python
df6.explain(True)
```

```python
Countrydistdf = df.select("Country").distinct()
Countrydistdf.show()
```

```python
Countrydistdf.explain(True)
```

```python
df7 = spark.range(100000)
df7.show()
df7.explain(True)
```

```python
df8 = spark.range(10000)
df9 = df8.selectExpr("id * 10 as id")
df9.explain(True)
```

```python
df10 = df8.repartition(6)
df10.explain(True)
```
