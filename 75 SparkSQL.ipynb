```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("readCsvFile").getOrCreate()
df = spark.read.csv("OrdersNew.csv", header=True, inferSchema=True)
df.show()
```

```python
df.createOrReplaceTempView("vwProducts")
```

```python

result = spark.sql("select * from vwProducts")
result.show()
result.count()
```

```python

result1 = spark.sql("select * from vwProducts where  Country = 'Italy'")
result1.show()
result1.count()
```

```python
df2 = spark.sql("SELECT Country, SUM(SalesAmount) AS TotalSales FROM vwProducts GROUP BY Country")
df2.show()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("readCsvFile").getOrCreate()
Customersdf = spark.read.csv("Customers.csv", header=True, inferSchema=True)
Customersdf.show()
```

```python
Salesdf.createOrReplaceTempView("vwSales")
Productsdf.createOrReplaceTempView("vwProducts")
Customersdf.createOrReplaceTempView("vwCustomers")
```

```python
joindf = spark.sql("""
SELECT v.ProductKey, v.SalesAmount, 
       p.ProductKey, p.EnglishProductName, 
       c.CustomerKey, c.CustomerName
FROM vwSales AS v
JOIN vwProducts AS p ON v.ProductKey = p.ProductKey
JOIN vwCustomers AS c ON v.CustomerKey = c.CustomerKey
""")

joindf.show()
```
